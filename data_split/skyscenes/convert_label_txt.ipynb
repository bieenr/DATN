```python
import os

def convert(root_dir, output_dir, dirpath=''):
    assert os.path.exists(root_dir), f"{root_dir} does not exist"
    
    with open(root_dir, 'r') as f:
        lines = f.readlines()
    
    # Tạo thư mục output_dir nếu chưa tồn tại
    os.makedirs(os.path.dirname(output_dir), exist_ok=True)
    
    with open(output_dir, 'w') as out_f:
        for line in lines:
            normal_path = line.strip()
            image_path = normal_path.replace('Normal', 'Images').replace('_normal', "")
            depth_path = normal_path.replace('Normal', 'Depth').replace('_normal', "_depth")
            assert os.path.exists(image_path), f"{image_path} does not exist"
            assert os.path.exists(depth_path), f"{depth_path} does not exist"
            image_path = image_path.split('/mnt/data/bientd/SkyScenes/')[1]
            depth_path = depth_path.split('/mnt/data/bientd/SkyScenes/')[1]
            out_f.write(f'{image_path} {depth_path}\n')

# Ví dụ gọi hàm
convert('/mnt/data/bientd/Marigold/data_split/skyscenes/train.txt', '/mnt/data/bientd/Marigold/data_split/skyscenes/skyscenes_train.txt')
# Ví dụ gọi hàm
convert('/mnt/data/bientd/Marigold/data_split/skyscenes/test.txt', '/mnt/data/bientd/Marigold/data_split/skyscenes/skyscenes_test.txt')
```
